```python
#최고기온 구하기
import csv

max_temp = -999
max_date = ''

f=open('C:\jupiterworkspace\pandas\data/weather_daegu.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
for row in data:
    if row[3] == '' :
        row[3] = -999
    if row[4] == '' :
        row[4] = -999
    if row[6] == '' :
        row[6] = -999
    if row[8] == '' :
        row[8] = -999

    row[3] = float(row[3])
    row[4] = float(row[4])
    row[6] = float(row[6])
    row[8] = float(row[8])
    
    if max_temp < row[4] :
        max_date = row[2]
        max_temp = row[4]
    
f.close()

print('2019~2021 3개년 중 가장 더웠던 날은 ', max_date, '이고, 그 날의 최고기온은 ', max_temp, '입니다.')
```

```python
import csv
import math
import matplotlib.pyplot as plt

f=open('C:\jupiterworkspace\pandas\data/gender.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

male = []
female = []
size=[]

gu = input('조사하고 싶은 구: ')

for row in data :

    if gu in row[0]:
        for i in range(11) :
            male.append(int(row[i+3].replace(",", "")))
            female.append(int(row[i+16].replace(",", "")))
            size.append(math.sqrt( int(row[i+3].replace(",", "")) + int(row[i+16].replace(",", "")) ))
        break

f.close()

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,5))
plt.title(gu + '의 성별 인구 그래프')
plt.scatter(male, female, s=size, c=range(11), alpha=0.5, cmap='jet') #컬러맵 적용
plt.colorbar()
plt.plot(range(max(male)), range(max(male)), 'g') #추세선 추가
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```

```python
#유임승차가 가장 많은 지하철역 찾기
import csv
import matplotlib.pyplot as plt

f=open('C:\jupiterworkspace\pandas\data/subwayfee.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

mx = 0
rate = 0
mx_station=''

for row in data :

    for i in range(4, 8) :
        row[i] = int(row[i])
    if row[6] != 0 and (row[4]+row[6]) >100000 : #10만 이하의 값은 너무 적어서 유의미한 통계수치를 내지 못한다고 판단
        rate = row[4]/(row[4]+row[6])
        if rate > mx :
            mx = rate
            mx_station = row[3] + ' ' + row[1]

f.close()

print(mx_station, round(mx*100, 2))
```

```python
#시간대별로 승차, 하차 인원이 가장 많은 역 찾기
import csv
import matplotlib.pyplot as plt

f=open('C:\jupiterworkspace\pandas\data/subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

enter_mx = [0]*24
enter_mx_station = ['']*24
out_mx = [0]*24
out_mx_station = ['']*24

for row in data :

    row[4:] = map(int, row[4: ])

    for j in range(24) :
        enter = row[2*j+4]
        out = row[2*j+5]

        if enter > enter_mx[j] :
            enter_mx[j] = enter
            enter_mx_station[j] = row[3] + '(' + str(j+4) + ')'

        if out > out_mx[j] :
            out_mx[j] = out
            out_mx_station[j] = row[3] + '(' + str(j+4) + ')'

f.close()

plt.rc('font', family='Malgun Gothic')

fig = plt.figure(figsize=(40,25))

axes1 = fig.add_subplot(2,1,1)
axes2 = fig.add_subplot(2,1,2)

axes1.bar(out_mx_station, out_mx, color='b')
axes2.bar(enter_mx_station, enter_mx, color='r')

axes1.set_title('시간대별 하차인원이 많은 역', size=30)
axes2.set_title('시간대별 승차인원이 많은 역', size=30)

axes1.set_xticklabels(out_mx_station, rotation=45, size=20)
axes2.set_xticklabels(enter_mx_station, rotation=45, size=20)

fig.tight_layout()
plt.show()
```
