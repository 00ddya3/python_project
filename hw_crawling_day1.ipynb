```python
#과제3 솔루션

import requests
from bs4 import BeautifulSoup
from openpyxl import Workbook

url = 'https://finance.naver.com/research/'

response = requests.get(url)
response.raise_for_status()
# Response에서 에러가 발생할 때
# 프로그램을 중단하도록 할 때는
# Response 객체의 raise_for_status() 메서드를 호출하여 코드를 불러옴

html = response.text

soup = BeautifulSoup(html, 'html.parser')

# #우클릭 > 복사 > selector 선택
# -> #contentarea_left > div:nth-child(1) > table
# 여기서 "div:nth-child(1)" 부분 해석을 위해
# "#contentarea_left " 부터 단계별 추적해 들어가야 함

pos = '#contentarea_left'
tbody = soup.select_one(pos)

print(tbody)
# "<div id="contentarea_left">" 하단 내용을 추출함
# 하단 첫번째 나오는 "<div class="box_type_m" ..> 블록에 "종목분석 리포트"가 담겨 있으
# 이 부분을 추출

div = tbody.select_one('div')
print(div)
# "종목분석 리포트" 내용이 담긴게 보인다.
# 분석해 보면
# <table summary="종목분석 리포트 리스트" ...> 하단에 내용이 있다.

# 첫번쩨 나오는 <table..> 태그이므로 select_one을 사용
table = div.select_one('table')
print(table)
# 추출된 내용을 보면 첫번째 <tr> 아래에 표의 테이블 제목이 나온다.
# <tr>
# <th>기업</th>
# <th>제목</th>
# <th class="tl">증권사</th>
# <th>첨부</th>
# <th>작성일</th>
# </tr>

# 타이틀을 추출해 보자
# <tr> 아래 컬럼 수 만큼 <th>가 있다.

tr = table.select_one('tr') # 제일 앞에 있는 것 선택하므로 select_one 사용

print(tr) # 타이틀이 보인다.

th = tr.select('th') # <th>를 다가져 올 것이므로 select() 사용

print('-' * 50)
print(th[0].text,"| ", th[1].text,"| ", th[2].text,"| ", th[3].text,"| ", th[4].text)

# 타이틀을 출력했다.
# 이제 내용을 가져오자.
# <table..> 하부에 있으니 찾자
print(table)
# 또 다른 <tr> 블록에 담겨있다.
# 그런데 <tr>은 이전에도 블록이 있다.

# <tr> 블록이 여러개 있음.
# 여기에는 타이틀이 담긴 <tr> 도 있다.
trs = table.select('tr')
print(trs)
# 리스트에 <tr> 기준으로 담겨있다

# <tr> 블록을 하나씩 빼보자
print(trs[0].text) # 타이틀이다.
print(trs[1].text) # 아무것도 없다.
# print(trs[2].text) # 2 부터 있고
print(trs[10].text) # 8,9는 공란이고, 10에서는 에러가 나온다.
# 3번째 <tr>에서 리포트 첫번째 라인 정보가 나온다.(기업,제목,증권사,작성일 이 있다.)
# 이것 분석하자.
# 그리고 7번 인덱스까지 정보가 있다. 여기까지만 처리하자.

print(trs[2])
# text는 <td>에 나누어져 담겨 있다.
# <td>를 담자

tds = trs[2].select('td')
for td in tds:
    print(td)
# 첫번째 <td>에 기업정보가
# 두번째 <td>에 제목이 있고
# 세번째 <td>에 증권사가 있고
# 네번째 <td>의 <a> 태그 속성 href 에 첨부파일 링크가
# 다섯번째 <td>에 날짜가 있다.
# --> 근데 앞에 공백이 많다. 없애자.
# --> 22.02.03으로 년도가 두자리다. 4자리로 만들자.


print(tds[0].text,tds[1].text,tds[2].text,tds[3].text,tds[4].text)
# 너무 붙어 있다. 길이를 넣고(.ljust(20)), 날짜는 조절하자(strip())
print('-' * 80)

# 첨부파일의 링크 가져오기
a = tds[3].find('a')
href = a['href'] # href 속성에 링크 정보 있음
# print(href)
# 날짜 조절하기
dt = '20' + tds[4].text.ljust(15).strip()
# print('변경된 Date: ', dt, "\n")
print(tds[0].text.ljust(10),tds[1].text.ljust(15),tds[2].text.ljust(15),dt, href.ljust(30))
# 내용 다 출력했다.
# 이제 반복해서 출력하자.

# 3번째 부터 마지막 까지가 내용이니 이 부분만 적용하자
idx = -1 # 반복문에 인덱스 번호를 0 부터 시작되게 하기 위해서
for tr in trs:
    idx += 1 # 인덱스 증가
    
    if idx <= 1 or idx > 7:
        continue

    tds = tr.select('td')
    title = tds[1].text.strip() # 제목 공백제거
    a = tds[3].find('a') # 첨부파일의 링크 가져오기
    href = a['href'] # href 속성에서 링크 정보 있음
    dt = '20' + tds[4].text.ljust(15).strip() # 날짜 조절하기
    print(tds[0].text.ljust(10),title.ljust(30),tds[2].text.ljust(15),dt, href.ljust(30))

```

```python
#과제 4

url = 'https://finance.naver.com/marketindex/exchangeList.naver'

response = requests.get(url)
response.raise_for_status()
# Response에서 에러가 발생할 때
# 프로그램을 중단하도록 할 때는
# Response 객체의 raise_for_status() 메서드를 호출하여 코드를 불러옴

if response.status_code == 200: # 정상
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    #우클릭 > 복사 > selector 선택
    #body > div > table > tbody
    tbody= soup.select_one('body > div > table > tbody')
    #print(soup)

    trs = tbody.find_all('tr')
else :
    print(response.status_code)

#파싱 + 출력
for tr in trs :
    tds = tr.find_all('td')
    print(tds[0].text.strip())
    print('매매기준율 :', '\t', tds[1].text)
    print('현찰(사실때) :', '\t',tds[2].text)
    print('현찰(파실때) :', '\t',tds[3].text)
    print('송금(보내실때) : ',tds[4].text)
    print('송금(받으실때) : ',tds[5].text)
    print('미화환산율 : ', '\t',tds[6].text)
    print('')
```
