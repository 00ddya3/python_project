```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def draw_barchart(citya_keylist, citya_valuelist, cityb_valuelist):
    x_labels = citya_keylist

    index = np.arange(len(x_labels))
    width = 0.4

    fig, ax = plt.subplots()
    rect1 = ax.bar(index - width/2, citya_valuelist, width, label="City A")
    rect2 = ax.bar(index + width/2, cityb_valuelist, width, label="City B")

    ax.set_xlabel('age')
    ax.set_ylabel('Population')                     # y축 레이블
    ax.set_title("The demographics of two cities")  # 그래프 타이틀

    plt.xticks(index + (width/10), x_labels)                   # X축 데이터
    plt.legend(fontsize=9, loc="upper right")
    fig

def draw_piechart(citya_total, citya_value, cityb_total, cityb_value, title_name):
    '''
    Pic chart를 그려줌
    citya_total: City A's total population
    citya_value: City A's a specific value

    cityb_total: City B's total population
    cityb_value: City B's a specific value
    '''
    citya_list = [citya_value, citya_total-citya_value]
    cityb_list = [cityb_value, cityb_total-cityb_value]
    label_list = [title_name , ""]

    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # City A의 pie chart
    ax1.pie(citya_list, labels=label_list, autopct="%.1f%%")
    ax1.set_title("City A's " +  title_name)

    # City B의 pie chart
    ax2.pie(cityb_list, labels=label_list, autopct="%.1f%%")
    ax2.set_title("City B's " + title_name)
    fig
```

```python

population = pd.DataFrame( 
    data={'0-9세' : [80, 120], '10-19세' : [100, 200], '20-29세' : [140, 300],
        '30-39세' : [160, 360], '40-49세' : [200, 400], '50-59세' : [240, 300],
        '60-69세' : [195, 40], '70-79세' : [160, 25], '80-89세' : [80, 20],
        '90-99세' : [20, 10]},
    index = ['CityA', 'CityB']
    )
print(population)
```

```python
#1
citya = population.loc['CityA']
cityb = population.loc['CityB']
citya_keylist=citya.keys().tolist()
citya_valuelist=citya.values.tolist()
cityb_valuelist=cityb.values.tolist()

print('Key: ', citya_keylist)
print('City A: ', citya_valuelist)
print('City B: ', cityb_valuelist)

draw_barchart(citya_keylist, citya_valuelist, cityb_valuelist)
```

```python
#2
citya_total = citya.sum()
cityb_total = cityb.sum()
citya_value = sum(citya_valuelist[2:10])
cityb_value = sum(cityb_valuelist[2:10])
title_name='Voting rate'

draw_piechart(citya_total, citya_value, cityb_total, cityb_value, title_name)

print('City A voting number:', citya_value, end="  ")
print('total number:', citya_total, end="  ")
print('voting rate:', round(citya_value/citya_total*100, 1), '%')

print('City B voting number:', cityb_value, end="  ")
print('total number:', cityb_total, end="  ")
print('voting rate:', round(cityb_value/cityb_total*100, 1), '%')
```

```python
#3
citya_value = sum(citya_valuelist[7:10])
cityb_value = sum(cityb_valuelist[7:10])
title_name='Old people rate'

draw_piechart(citya_total, citya_value, cityb_total, cityb_value, title_name)

print('City A senior number:', citya_value, end="  ")
print('total number:', citya_total, end="  ")
print('voting rate:', round(citya_value/citya_total*100, 1), '%')

print('City B senior number:', cityb_value, end="  ")
print('total number:', cityb_total, end="  ")
print('voting rate:', round(cityb_value/cityb_total*100, 1), '%')
```
