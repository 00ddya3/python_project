```python
url = 'https://finance.naver.com/'

response = requests.get(url)
response.raise_for_status()
# Response에서 에러가 발생할 때
# 프로그램을 중단하도록 할 때는
# Response 객체의 raise_for_status() 메서드를 호출하여 코드를 불러옴

if response.status_code == 200: # 정상
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    #우클릭 > 복사 > selector 선택
    #container > div.aside > div > div.aside_area.aside_popular > table > tbody (인기검색종목)
    tbody = soup.select_one('#container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody')
    #print(soup)

    trs = tbody.select('tr') # <tr> 영역 배열에 넣기
    print(trs)
else :
    print(response.status_code)
```

```python
search_word = input('검색어를 입력하세요 : ')
start_pg = int(input('가져올 첫 페이지를 입력하세요: '))
last_pg = int(input('가져올 마지막 페이지를 입력하세요: '))
#print('[검색어] :' + search_word + '[검색페이지] :' + str(page))


#https://search.naver.com/search.naver?where=news&sm=tab_jum&query=코로나&start=1
base_url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&"
page_num = last_pg - start_pg + 1


page = start_pg
for i in range(page_num) :
    print('페이지번호: ', page)

    #url 생성
    web_page = (page - 1) * 10 + 1
    url = base_url + "query=" + str(search_word) + "&start=" + str(web_page)

    #url 가져오기 오류 보정
    response = requests.get(url)
    if response.status_code == 200 : #정상
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
    else :
        print(response.status_code)

    #기사 타이틀 출력
    # select선택 ==> #sp_nws29 > div.news_wrap.api_ani_send > div > a
    # div class="news_wrap api_ani_send"인데 중간에 공란이 있어서 앞부분만 "div.news_wrap"
    #sp_nws1 > div.news_wrap.api_ani_send > div > a 
    # #nws 뒤의 넘버는 기사마다 다름
    pos = "div.group_news > ul.list_news > li.bx > div.news_wrap > div.news_area > a"
    divs = soup.select(pos) # 리스트 출력

    # 링크 추출 title['href']

    #기사 신문사 추출
    #sp_nws86 > div > div > div.news_info > div.info_group > a.info.press > span


    for j, title in enumerate(divs) : 
        print(j, '\t', '기사제목: ', title.text, '\n\t url: ', title['href'])

    page += 1
```
