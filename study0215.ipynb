```python
#나이가 50살이상이고, 여자인 승객의 명단을 출력하세요.
con1 = df['age'] >= 50
con2 = df['sex'] == 'female'
df[con1 & con2]
```

```python
#항구(embarked) 'S'에서 탑승한 승객 중 살아남은 사람은 몇명인가?
con1 = df['survived']==1
con2 = df['embarked']=='S'
print(df[con1 & con2].shape[0])
```

```python
df.mode()
```

```python
df.isnull().sum() #열 별 결측치 수
df.isnull().sum().sum() #전체 결측치 수

df['age'] = df['age'].fillna(round(df['age'].mean(), 1)) #age의 결측치를 평균값으로 채우기
df.tail()

df['deck'] = df['deck'].fillna(df['deck'].mode()[0]) #deck의 결측치를 최빈값으로 채우기
df.tail()
```
