```python
#전체 코드


import requests
from urllib.parse import urlparse
import csv

def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/blog?query=" + \
        keyword + \
        "&display=" + str(display_count) + \
        "&start=" + str(start)
    result = requests.get(urlparse(url).geturl(),
        headers={"X-Naver-Client-Id":"Fid_LQiP9j9aM7e4KWpE", "X-Naver-Client-Secret":"6IxM7NZWFK"})
    return result.json()

def call_and_print(keyword, display, page):
    json_obj = get_api_result(keyword, display, start_point)
    for item in json_obj['items']:
        title = item['title'].replace("<b>","").replace("</b>","")
        print(title + "@" + item["bloggername"] + "@"
            + item["postdate"] + "@" + item["link"] )

def call_and_csv_write(keyword, display, page):
    json_obj = get_api_result(keyword, display, start_point)
    for item in json_obj['items']:
        title = item['title'].replace("<b>","").replace("</b>","")
        wr.writerow([title, item["bloggername"],item["postdate"], item["link"] ])

#___ 변수에 값 배정 ___
keyword = "강남역"
display_count = 100
start_point = 1
repeat_times = 5
#_______________________

#======== 출력하기 =======
for cnt in range(repeat_times):
    print('--- { count } ---------', cnt)
    call_and_print(keyword,display_count,start_point)
    start_point += display_count

json_obj = get_api_result(keyword, display_count, start_point)

#================ csv로 저장 하기 ==========
f = open('naver_api.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(f)

keyword = "강남역"
display_count = 100
start_point = 1

# 반복처리 _____
for cnt in range(repeat_times):
    # print('--- { cnt} ---------', cnt)
    call_and_csv_write(keyword,display_count,start_point)
    start_point += display_count
f.close()

# --- 다시 읽어서 건수 확인하기
f = open('naver_api.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
cnt = 0
# print(len(rdr))
for line in rdr:
    cnt += 1
# print(line)

print('count: ', cnt)
f.close()
```
