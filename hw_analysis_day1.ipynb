```python
import pandas as pd
import seaborn as sns

df=sns.load_dataset('titanic')
```

```python
#탑승 항구 별 승객 데이터 분포
df['embark_town'].value_counts()
```

```python
#요금범위 별 승객 데이터 분포
df['fare'].value_counts().sort_index()
```

```python
#클래스 별 승객 데이터 분포
df['class'].value_counts()
```

```python
#who별 승객 데이터 분포
df['who'].value_counts()
```

```python
#나이가 50살이상이고, 여자인 승객의 명단을 출력하세요.
con1 = df['age'] >= 50
con2 = df['sex'] == 'female'
df[con1 & con2]
```

```python
#항구(embarked) 'S'에서 탑승한 승객 중 살아남은 사람은 몇명인가?
con1 = df['survived']==1
con2 = df['embarked']=='S'
print('S 탑승자 중 생존자 수:', df[con1 & con2].shape[0])
```

```python
#3등석 승객 중 여자는 몇명 인가? 그 중 생존자는 몇명인가?
con1 = df['pclass'] == 3
con2 = df['sex'] == 'female'
con3 = df['survived'] == 1

print('3등석 여성 수:', df[con1 & con2].shape[0])
print('3등석 여성 생존자 수:', df[con1 & con2 & con3].shape[0])
```

```python
#embark_town이 'Queenstown'인 데이터만 추출
con = df['embark_town'] == 'Queenstown'
df.loc[con]
```

```python
#남자 아이 데이터만 추출(sex=male, who=child)
con1 = df['sex'] == 'male'
con2 = df['who'] == 'child'
df.loc[con1 & con2]
```

```python
#요금(fare)을 40 이상 50 미만 금액을 지불한 승객의 평균나이와 승객수
con1 = df['fare'] >= 40
con2 = df['fare'] < 50
df_fare = df[con1 & con2]
print('40이상 50미만 요금을 낸 승객의 평균나이:', round(df_fare['age'].mean(), 1) )
print('40이상 50미만 요금을 낸 승객의 수', df_fare.shape[0])
```

```python
#2등급(pclass) 승객의 평균운임, 평균나이와 승객수
df_class2 = df[df['pclass']==2]
print('2등급 승객의 평균운임:', round(df_class2['fare'].mean(), 2))
print('2등급 승객의 평균나이:', round(df_class2['age'].mean(), 1))
print('2등급 승객 수:', df_class2.shape[0])
```

```python
#위 두 조건을 동시에 만족하는 승객수와 승객의 평균나이는
df_inner = pd.concat([df_fare, df_class2], join='inner', axis=1)
print('위의 두 조건을 만족하는 승객수:', df_inner.shape[0])
print('위의 두 조건을 만족하는 승객의 평균나이:', round(df_inner['age'].mean(), 1))
```

```python
#열 별 결측치 수, 전체 결측치 수
print('열 별 결측치 수:\n', df.isnull().sum())
print('전체 결측치 수:', df.isnull().sum().sum())
```

```python
#age의 결측치를 평균값으로 채우기
df['age'] = df['age'].fillna(round(df['age'].mean(), 1)) 
df.tail()
```

```python
#deck의 결측치를 최빈값으로 채우기
df['deck'] = df['deck'].fillna(df['deck'].mode()[0]) 
df.tail()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

titanic = sns. load_dataset('titanic')
flights = sns. load_dataset('flights')
```

```python
#데이터요소: x - who, y - survived
#그래프별 제목: 임의로 부여하세요.
#sect1: x,y 값만
#sect2: "hue: class" 추가
#sect3: sect2를 누적그래프로

sns. set_style('darkgrid') # 스타일 설정(dark, darkgrid, white,whithgrid,ticks)

# 그래프 객체 생성(도화지 한장 만든 것임)
fig = plt. figure(figsize= (10,5))

# 도화지 분할(add_subplot(행, 열, 인덱스))
sect1 = fig. add_subplot(1,3,1)
sect2 = fig. add_subplot(1,3,2)
sect3 = fig. add_subplot(1,3,3)

# 부분 섹터별 설정
sect1. set_title("who:servived")
sect2. set_title("who:servived_hue")
sect3. set_title("who:servived_dodge")

# 섹터에 그래프 그리기
# hue:범레, 데이터분류
# dodge: 항목별 누적여부(True:개별, False:누적)
sns. barplot(x= 'who', y= 'survived', data= titanic, ax= sect1)
sns. barplot(x= 'who', y= 'survived', hue= 'who', data= titanic, ax= sect2)
sns. barplot(x= 'who', y= 'survived', hue= 'who', dodge= False , data= titanic, ax= sect3)
```

```python
#데이터요소: x - embarked
#그래프별 제목: 임의로 부여하세요.
#sect1: x,y 값만
#sect2: "hue: class" 추가
#sect3: sect2를 누적그래프로

sns. set_style('whitegrid') # 스타일 설정(dark, darkgrid, white,whitegrid,ticks)

# 그래프 객체 생성(도화지 한장 만든 것임)
fig = plt. figure(figsize= (10,5))

# 도화지 분할(add_subplot(행, 열, 인덱스))
sect1 = fig. add_subplot(1,3,1)
sect2 = fig. add_subplot(1,3,2)
sect3 = fig. add_subplot(1,3,3)

# 부분 섹터별 설정
sect1. set_title("embarked:age")
sect2. set_title("embarked:fare_hue")
sect3. set_title("embarked:fare_dodge")

# 섹터에 그래프 그리기
# swarmplot: Draw a categorical scatterplot with non-overlapping points.
sns. stripplot(x= 'embarked', y= 'age', data= titanic, ax= sect1)
sns. stripplot(x= 'embarked', y= 'fare', hue= 'embarked', data= titanic, ax= sect2)
sns. swarmplot(x= 'embarked', y= 'fare', hue= 'embarked', dodge= False , data= titanic, ax= sect3)
```
